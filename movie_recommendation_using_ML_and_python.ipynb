```python
#read the csv files from the database using the file's path
credits_df = pd.read_csv("/tmdb_5000_credits.csv")
movies_df = pd.read_csv("/tmdb_5000_movies.csv")
```

```python
#print the movies dataframe
movies_df.head()
```

```python
#print the credits dataframe
credits_df.head()
```

```python
#merge the movie and credits dataframes

credits_df.columns = ['id','title','cast','crew']

movies_df = movies_df.merge(credits_df, on="id")
```

```python
#new dataframe after merging
movies_df.head()
```

```python
#applying literal_eval on each and every feature.
#Before applying literal_eval type of crew is List
#after applying literal_eval type of crew is dictionary
from ast import literal_eval
features = ["cast", "crew", "keywords", "genres"]
for feature in features:
    movies_df[feature] = movies_df[feature].apply(literal_eval)
movies_df[features].head(10)
```

```python
#get the name of the director of the movie.
def get_director(x):
    for i in x:
        if i["job"] == "Director":
            return i["name"]
    return np.nan
```

```python
#get the top 3 elements or the entire list whichever is more.
def get_list(x):
    if isinstance(x, list):
        names = [i["name"] for i in x]

        if len(names) > 3:
            names = names[:3]

        return names

    return []
```

```python
#apply both the functions get_director() and get_list() to our dataset
#apply the crew information to get_director
#extract the top 3 elements for keywords,genres,cast by applying get_list()
movies_df["director"] = movies_df["crew"].apply(get_director)

features = ["cast", "keywords", "genres"]
for feature in features:
    movies_df[feature] = movies_df[feature].apply(get_list)
```

```python
#print the data after above transformation
movies_df[['original_title', 'cast', 'director', 'keywords', 'genres']].head()
```

```python
#convert the above feature instances into lowercase and remove all the spaces between them
def clean_data(row):
    if isinstance(row, list):
        return [str.lower(i.replace(" ", "")) for i in row]
    else:
        if isinstance(row, str):
            return str.lower(row.replace(" ", ""))
        else:
            return ""

features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    movies_df[feature] = movies_df[feature].apply(clean_data)
```

```python
#create a “soup” containing all of the metadata information extracted to input into the vectorizer
def create_soup(features):
    return ' '.join(features['keywords']) + ' ' + ' '.join(features['cast']) + ' ' + features['director'] + ' ' + ' '.join(features['genres'])


movies_df["soup"] = movies_df.apply(create_soup, axis=1)
print(movies_df["soup"].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

#countVectorizer counts the frequency of each word and outputs a 2D vector containing frequencies
count_vectorizer = CountVectorizer(stop_words="english")
count_matrix = count_vectorizer.fit_transform(movies_df["soup"])

#print the shape of the count_matrix formed after vectorizing the obtained soup using countvectorizer
print(count_matrix.shape)

#cosine similarity is the dot product of the vector outputs by the CountVectorizer
cosine_sim2 = cosine_similarity(count_matrix, count_matrix) 
#print the cosine similarity shape after finding the cosine similarity between count_matrix nad count_matrix
print(cosine_sim2.shape)

#movies_df = movies_df.reset_index()
indices = pd.Series(movies_df.index, index=movies_df['original_title'])
```

```python
#create a reverse mapping of movie titles to indices

indices = pd.Series(movies_df.index, index=movies_df["original_title"]).drop_duplicates()

print(indices.head())
```

```python
#create a function that takes in the movie title and the cosine similarity score as input and outputs the movies similar to it
def get_recommendations(original_title, cosine_similarity=cosine_similarity):
    idx = indices[original_title]
    similarity_scores = list(enumerate(cosine_similarity[idx]))
    similarity_scores= sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similarity_scores= similarity_scores[1:11]
    # (a, b) where a is id of movie, b is similarity_scores

    movies_indices = [ind[0] for ind in similarity_scores]
    movies = movies_df["original_title"].iloc[movies_indices]
    return movies


print("################ Content Based System #############")
print("Recommendations for The Dark Knight Rises")
print(get_recommendations("The Dark Knight Rises", cosine_sim2))
print()
print("Recommendations for Avengers")
print(get_recommendations("The Avengers", cosine_sim2))
```
